```python
import pandas as pd
```

```python
dataset=pd.read_csv('netflix_titles.csv',index_col='show_id')
dataset
```

```python
dataset.drop(['description','cast','director'],axis=1,inplace=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
missing_values=dataset.isnull().sum()
missing_values
```

```python
dataset['country'].fillna('Unknown', inplace=True)
dataset['date_added'].fillna('Unknown', inplace=True)
dataset['duration'].fillna('Unknown', inplace=True)
dataset['rating'].fillna('Unknown', inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(data=dataset,x='type')
plt.title('movie vs Tv show')
plt.show()
```

```python
df=dataset[dataset['duration'].str.contains('min',na=False)].copy()
df['duration_mintues']=df['duration'].str.replace('min','').astype(int)
```

```python
sns.histplot(df['duration_mintues'],bins=20,kde=True)
plt.title('Distribution of movie duration')
plt.xlabel('Duration(minutes)')
plt.ylabel('frequency')
plt.show()
```

```python
top_countries = dataset['country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Top 10 Countries with Most Content')
plt.xlabel('Number of Titles')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset['rating']=le.fit_transform(dataset['rating'])
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='rating', y='duration_mintues', data=df)
plt.title('Rating vs Duration')
plt.xlabel('Rating')
plt.ylabel('Duration (minutes)')
plt.show()
```
